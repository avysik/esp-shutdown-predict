```python
import pandas as pd

path = "out_system_day/esp-system__esp_1006__run_00000.parquet"

df = pd.read_parquet(path)

print("shape:", df.shape)
print("\ncolumns:")
print(df.columns.tolist())

print("\ndtypes:")
print(df.dtypes)

print("\nhead:")
display(df.head())
```

```python
# Проверка временной оси
print("min timestamp:", df["value_date"].min())
print("max timestamp:", df["value_date"].max())

print("\nunique step:")
print(df["value_date"].diff().dropna().unique())

# Есть ли пропуски
print("\nNaNs per column:")
print(df.isna().sum())

# Простая статистика
print("\ndescribe:")
display(df.describe())
```
