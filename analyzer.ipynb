```python
import pandas as pd

path = "out_system_day/esp-system__esp_1006__run_00000/esp-system__esp_1006__run_00000.parquet"

df = pd.read_parquet(path)

print("shape:", df.shape)
print("\ncolumns:")
print(df.columns.tolist())

print("\ndtypes:")
print(df.dtypes)

print("\nhead:")
display(df.head())
```

```python
# Проверка временной оси
print("min timestamp:", df["value_date"].min())
print("max timestamp:", df["value_date"].max())

print("\nunique step:")
print(df["value_date"].diff().dropna().unique())

# Есть ли пропуски
print("\nNaNs per column:")
print(df.isna().sum())

# Простая статистика
print("\ndescribe:")
display(df.describe())
```

```python
from pathlib import Path
import matplotlib.pyplot as plt

base_dir = Path("out_system_day") # out_batch

for exp_dir in base_dir.glob("*"): # esp_* по всем
    if not exp_dir.is_dir():
        continue

    print(f"Processing {exp_dir.name}")

    for parquet_file in exp_dir.glob("*.parquet"):
        print(f"  File: {parquet_file.name}")

        df = pd.read_parquet(parquet_file)

        # берём только числовые колонки
        df = df.select_dtypes(include="number")

        for col in df.columns:
            plt.figure(figsize=(8, 4))
            df[col].plot(title=f"{exp_dir.name} | {parquet_file.name} | {col}")
            plt.grid(True)
            plt.tight_layout()
            plt.show()
```
